```python
%%writefile baie.cpp
#include <iostream>
using namespace std;

int findMax(int num1, int num2, int num3) {
    int max = num1;
    if (num2 > max) max = num2;
    if (num3 > max) max = num3;
    return max;
}

int main() {
    int a, b, c;
    cin >> a >> b >> c;
    cout << findMax(a, b, c) << endl;
    return 0;
}
```

```python
echo 2 3 5 > baie_t1.inp
echo 3 3 3 > baie_t2.inp
```

```python
!g++ baie.cpp -o baie.exe
!baie.exe < baie_t1.inp > out1.txt
!baie.exe < baie_t2.inp > out2.txt
!echo Expected: 5
!type out1.txt
!echo Expected: 3
!type out2.txt
```

```python
expected = ["5", "3"]
files = ["out1.txt", "out2.txt"]

for i, (exp, fname) in enumerate(zip(expected, files), 1):
    with open(fname) as f:
        got = f.read().strip()
    if got == exp:
        print(f"Test {i}: PASS (Expected={exp}, Got={got})")
    else:
        print(f"Test {i}: FAIL (Expected={exp}, Got={got})")
```
